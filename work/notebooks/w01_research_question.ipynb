## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What I can claim:

1. I can identify which pages are currently showing decline signals (low CTR, weak position, low visibility) based on observed data.
2. I can rank pages by risk — this is a decision-support tool to help a reviewer prioritize, not an automated fix.
3. My results are observational and directional: "pages with these patterns tend to decline," not proof of a hidden cause.

What I can't claim:

1. I can't claim a refresh causes recovery — that needs a real experiment (e.g., refresh some pages, leave others alone, compare results), which this data doesn't give me.
2. I can't claim to predict or explain Google's ranking algorithm — I'm only using observed signals (impressions, clicks, position), not Google's internal logic.
3. My current label (trend_direction == "down") is a snapshot, not a true future outcome — so today's results describe patterns in this window, not a validated forecast yet.
4. High precision on this dataset doesn't guarantee the same performance on new clients or the full warehouse — it needs to be re-validated at scale.
